To start with the Left component which is the Grid heat map where Columns are County and Rows are the governmental department (Agency Name). Here it is important to note that both of these columns/variables Agency Names and County are categorical variables and we need to plot these on a heat map with values for the mean of total square footage for that set of selected criteria. 
Since it involves 3D plotting where we plot Agency Names and County which needs to be selected to get values for the means of square footage for the selected Agency Name with respect to County I have used a pivot table for simplicity of plotting.
Pivot table will give us what we want in the form of a table which we will convert in a heat grid map going ahead.
I have used the link provide by Jill Naiman in slack to understand pivot table part of coding.
https://datagy.io/python-pivot-tables/ 
This link was really helpful.
Here index denotes rows which in our case is Agency Name, columns denotes County and Values denotes Square Footage. Since we want to find the mean of total Square Footage I have set the aggregation function to mean. I have cleaned the data by using dropna=True and fill_value=0.

```python
pivot1=pd.pivot_table(bldg, index="Agency Name", columns="County",values="Square Footage",aggfunc='mean',dropna=True,fill_value=0)
```

Here I have constructed a heat map. For this I have followed the steps taught by Jill Naiman in class. I have used the code from class and prep note book.
Step1 adding the 3 scales that is the color scale, x and y scales
Step2. Axis -- color & for x/y
Step3  Using heatmap to plot
Step4. building interactions
Step5 Adding interaction function
Step6 Linking  interaction function to the mark
Step7 Putting everything together in a figure
Step8 Using a Dashboard to display the interactivity

```python
# step1 to add scales - colors, x & y
col_sc = bqplot.ColorScale(scheme="Orange") 
x_sc = bqplot.OrdinalScale() 
y_sc = bqplot.OrdinalScale()

#step2. Axis -- color & for x/y
ax_col = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
ax_x = bqplot.Axis(scale=x_sc,label="County")
ax_y = bqplot.Axis(scale=y_sc, orientation='vertical',label="Agency Names")

# Step3. Using heatmap to plot
heat_map = bqplot.GridHeatMap(color=pivot1.values,row =pivot1.index,column =pivot1.columns,
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc}, interactions={'click':'select'}, 
                             anchor_style={'fill':'blue'},unselected_style = {'opacity': 2.0})

# step4. building interactions
mySelectedLabel = ipywidgets.Label()
mySelectedLabel.value = 'Mean selected point here'
# step5 interaction function -- this is what happens on selection:
def on_selected(change):
    #print(change)
    if len(change['owner'].selected) == 1: # this will allow the user to select one bin
        i,j = change['owner'].selected[0] # the only selected bin
        v = pivot1.values[i,j] #passing pivot1.values (mean square footage) for selected i and j (Agency Value and County)
        mySelectedLabel.value = 'Mean data value= '+ str(v)
# step 6 for linking  interaction function to the mark:
heat_map.observe(on_selected, 'selected')

# step7 Finally, putting together in a figure
fig_heatmap= bqplot.Figure(marks=[heat_map], axes=[ax_col, ax_x, ax_y])

# step8 to display on a dashboard.
myDashboard = ipywidgets.VBox([mySelectedLabel, fig_heatmap])
myDashboard
```

For the Final Dashboard to link the heat grid map with the bar plot I have added another step. Step 4 defines a function #in_barplot(a,b). This function will basically take input values from step 7 in_barplot(pivot1.index[i],pivot1.columns[j]) and calculate the output .groupby("Year Acquired")["Square Footage"].sum() to be returned for the corresponding Agency name and County from the main dataset building. This will help us add interactivity between the heat map and bar plot shown in step 7.
For this I read a lot of related documents from google. I tried many combinations of the code to get the final answer.
Finally to display the dashboard I used Jill Naiman's class code. If you click on any of the grid of the heat map you will get the corresponding year acquired and total square footage of the years in the bar plot. 
I have displayed the dashboard vertically for more clarity as using the Horizontal Box looked very messy in my case.

```python
# step1 to add scales - colors, x & y
col_sc = bqplot.ColorScale(scheme="Orange") 
x_sc = bqplot.OrdinalScale() 
y_sc = bqplot.OrdinalScale()

#step2. Axis -- color & for x/y
ax_col = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
ax_x = bqplot.Axis(scale=x_sc,label="County")
ax_y = bqplot.Axis(scale=y_sc, orientation='vertical',label="Agency Names")

# Step3. Using heatmap to plot
heat_map = bqplot.GridHeatMap(color=pivot1.values,row =pivot1.index,column =pivot1.columns,
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc}, interactions={'click':'select'},
                        anchor_style={'fill':'blue'},unselected_style = {'opacity': 2.0})

#step4 Defined a function for the bar plot
def in_barplot(a,b):
    val=bldg[(bldg["Agency Name"]==a) & (bldg["County"]==b)]
    val = val[val['Year Acquired'] != 0] 
    val=val.groupby("Year Acquired")["Square Footage"].sum()
    return val

# step5. building interactions
mySelectedLabel = ipywidgets.Label()
mySelectedLabel.value = 'Mean selected point here'
# step6 interaction function -- this is what happens on selection:
def on_selected(change):
    #print(change)
    if len(change['owner'].selected) == 1: # this will allow the user to select one bin
        i,j = change['owner'].selected[0] # the only selected bin
        v = pivot1.values[i,j] #passing pivot1.values (mean square footage) for selected i and j (Agency Value and County)
        mySelectedLabel.value = 'Mean data value= '+ str(v)
# step 7 for linking  interaction function to the mark:
        bldg_final=in_barplot(pivot1.index[i],pivot1.columns[j])
    
        bar_to_plot.x=bldg_final.index
        bar_to_plot.y=bldg_final.values
heat_map.observe(on_selected, 'selected')

# step8 Finally, putting together in a figure
fig_heatmap = bqplot.Figure(marks=[heat_map], axes=[ax_col, ax_x, ax_y])
fig_barplot = bqplot.Figure(marks=[bar_to_plot], axes=[x_axh, y_axh])

# step9 to display on a dashboard.
myDashboard = ipywidgets.VBox([mySelectedLabel, ipywidgets.VBox([fig_heatmap,fig_barplot])])
myDashboard
```
